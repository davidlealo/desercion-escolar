```python
# Kprototypes de desertores2018

from kmodes.kprototypes import KPrototypes
from sklearn.preprocessing import StandardScaler

# Crear matriz de datos
data = desertores2018.iloc[:, 1:].values

# Crear matriz de datos categóricos
data_cat = desertores2018.iloc[:, 1:5].values

# Crear matriz de datos numéricos
data_num = desertores2018.iloc[:, 5:].values

# Crear objeto de la clase StandardScaler
sc = StandardScaler()

# Ajustar y transformar los datos numéricos
data_num = sc.fit_transform(data_num)

# Crear objeto de la clase KPrototypes
kproto = KPrototypes(n_clusters=5, init='Cao')

# Ajustar el modelo a los datos
kproto.fit(data, categorical=[0, 1, 2, 3])

# Obtener las etiquetas de los clusters
clusters = kproto.labels_

# Obtener los centroides
centroids = kproto.cluster_centroids_

# Visualizar los centroides
print(centroids)
```
