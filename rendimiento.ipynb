```python
# Random Forest d2018   
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.preprocessing import StandardScaler
```

```python

# Dividir datos en entrenamiento y prueba
X = df2018.drop('DESERTOR', axis=1)
y = df2018['DESERTOR']
X_traindf, X_testdf, y_traindf, y_testdf = train_test_split(X, y, test_size=0.3)
```

```python
X_traindf.head()
```

```python
# Escalar datos
scaler = StandardScaler()
scaler.fit(X_traindf)
X_traindf = scaler.transform(X_traindf)
X_testdf = scaler.transform(X_testdf)
```

```python
# Entrenar modelo
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(X_traindf, y_traindf)

# Predecir
y_preddf = rfc.predict(X_testdf)

# Evaluar modelo
print(confusion_matrix(y_testdf, y_preddf))
print(classification_report(y_testdf, y_preddf))

# Importancia de las variables
feature_imp = pd.Series(rfc.feature_importances_, index=X.columns).sort_values(ascending=False)
print(feature_imp)

# Visualizar importancia de las variables
sns.barplot(x=feature_imp, y=feature_imp.index)
plt.xlabel('Importancia de las variables')
plt.ylabel('Variables')
plt.title('Visualizacion de importancia de las variables')
plt.show()
```
