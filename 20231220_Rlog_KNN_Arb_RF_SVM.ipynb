```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/tablas/seguimiento1stsem2018.csv', sep=';')
```

# Mapa de correlación

```python
df.head()
```

```python
df.info()
```

```python
# Datos nulos
df.isnull().sum()
```

```python
# Drop AGNO y COD_MEN
df = df.drop(['AGNO', 'COD_MEN'], axis=1)
```

```python
df.info()
```

```python
df = df.drop(['NOM_RBD'], axis=1)
```

```python
df = df.drop(['NOM_REG_RBD_A'], axis=1)
```

```python
df = df.drop(['NOM_COM_RBD'], axis=1)
```

```python
df = df.drop(['NOM_DEPROV_RBD'], axis=1)
```

```python
# columnas object
df.select_dtypes(include=['object']).columns
```

```python
df = df.drop(['LET_CUR', 'NOM_COM_ALU', 'SIT_FIN'], axis=1)
```

```python
# Convertir a dummies la variable SIT_FIN_R
df = pd.get_dummies(df, columns=['SIT_FIN_R'])
```

```python
df['DESERTOR'].value_counts()
```

```python
# Drop variables: COD_REG_RBD, COD_PRO_RBD, COD_DEPROV_RBD, COD_REG_ALU, COD_COM_ALU, NOM_COM_ALU, COD_RAMA, COD_SEC, COD_ESPE
df = df.drop(['COD_REG_RBD', 'COD_PRO_RBD', 'COD_DEPROV_RBD', 'COD_REG_ALU', 'COD_COM_ALU', 'COD_RAMA', 'COD_SEC', 'COD_ESPE'], axis=1)
```

```python
corr_data = df.corr()
corr_data
```

```python
corr_data = df.corr()
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr_data, cmap = cmap)
```

```python
# Dividir X y y con variable DESERTOR en y y escalar datos
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

X = df.drop(['DESERTOR'], axis=1)
y = df['DESERTOR']

scaler = StandardScaler()
X = scaler.fit_transform(X)

# Dividir datos en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y,random_state=2023)
```

```python
# regresion logistica
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Predicciones
y_pred = logreg.predict(X_test)

# Matriz de confusion
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
# Grafico de matriz de confusion
ax = plt.subplot()
sns.heatmap(confusion_matrix, annot=True, ax=ax, cmap='Blues', fmt='g')
ax.set_xlabel('Predicted labels')
ax.set_ylabel('True labels')
ax.set_title('Confusion Matrix')
ax.xaxis.set_ticklabels(['No desertor', 'Desertor'])
ax.yaxis.set_ticklabels(['No desertor', 'Desertor'])
plt.show()
```

```python
# Accuracy
from sklearn.metrics import accuracy_score
print(f'Accuracy {accuracy_score(y_test, y_pred)}')

# Precision
from sklearn.metrics import precision_score
precision_score(y_test, y_pred)
print(f'Precision {precision_score(y_test, y_pred)}')

# Recall
from sklearn.metrics import recall_score
recall_score(y_test, y_pred)
print(f'Recall {recall_score(y_test, y_pred)}')

# F1
from sklearn.metrics import f1_score
f1_score(y_test, y_pred)
print(f'F1 {f1_score(y_test, y_pred)}')
```

```python
# Curva ROC
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure(figsize=(10,6))
plt.plot(fpr, tpr, label='Regresion Logistica (area = %0.2f)' % logit_roc_auc)
plt.plot([0,1], [0,1], 'r--')
plt.xlim([0,1])
plt.ylim([0,1.05])
plt.xlabel('Tasa de Falsos Positivos')
plt.ylabel('Tasa de Verdaderos Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.savefig('Log_ROC')
plt.show()
```

```python
# Variables importantes
importance = logreg.coef_[0]
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))

# Grafico de variables importantes
plt.figure(figsize=(10,6))
plt.bar([x for x in range(len(importance))], importance)
plt.xticks(np.arange(0, 20, step=1))
plt.xlabel('Variables')
plt.ylabel('Importancia')
plt.title('Variables importantes')
plt.show()
```

```python
# Convertir importance a nombres de variables
importance = pd.DataFrame(importance)
importance = importance.T
importance.columns = df.drop(['DESERTOR'], axis=1).columns
importance = importance.T
importance.columns = ['Importancia']
importance = importance.sort_values(by=['Importancia'], ascending=False)
importance
```

```python
# Grafico de variables importantes
plt.figure(figsize=(10,6))
plt.bar(importance.index, importance['Importancia'])
plt.xticks(rotation=90)
plt.xlabel('Variables')
plt.ylabel('Importancia')
plt.title('Variables importantes')
plt.show()
```

```python
# Variables con importancia mayor a 0.099011
importance = importance[importance['Importancia'] > 0.099011]
importance
```
