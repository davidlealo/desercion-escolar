```python
from kmodes.kprototypes import KPrototypes

# Definir el número de clusters
num_clusters = 3

# Crear una instancia de KPrototypes
kproto = KPrototypes(n_clusters=num_clusters, init='Cao', verbose=2)

# Ajustar el modelo a los datos
clusters = kproto.fit_predict(df, categorical=[9, 10, 14, 15, 16, 18, 23, 24, 28, 29, 30, 31, 32, 33, 34, 35, 36])

# Añadir las etiquetas de los clusters al dataframe
df['Cluster'] = clusters
```
